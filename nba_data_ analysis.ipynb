```python
#games details datasına oyuncular üçlük atış başarı oranına göre "3_pts_success" adında yeni bir sütun eklenmiştir. Oyuncunun o maç için üçlük atış başarı oranı 0.5 ve daha fazlaysa o satırda Excellent, 0.5-0.3 arası ise Good ve daha düşükse Not good yazacaktır.
def three_point(rate):
    if rate >= 0.5:
        return "Excellent"
    elif rate < 0.5 and rate >= 0.3:
         return "Good"
    elif rate < 0.3:
         return "Not good"
        
df1["3_pts_success"] = df1.fg3_pct.apply(three_point)
```

```python
# fp = fancy player -> oyuncunun o maç içerisinde aldığı puanı, rebound, assist, top çalma ve blok sayısını toplayarak bu değerden turnover ve faul sayısını çıkartarak oyuncuların performanslarını ölçmek amacıyla hesaplanmıştır. Bir sütun olarak df1'e eklenmiştir.
df1["fp"] = df1["pts"] + df1["reb"] + df1["ast"]+ df1["stl"] + df1["blk"] - df1["to"] - df1["pf"]
```

```python
#Takımların ev sahibi ve deplasman olarak maçlarda aldıkları puana göre kategori edildiği home ve away sütunlarının eklenmesi
def func(pts):
    if pts > 100:
        return "Perfect"
    elif pts <= 100 and pts >= 80:
        return "Good"
    elif pts < 80:
        return "Average"
    
df2["home"] = df2.pts_home.apply(func)
df2["away"] =df2.pts_away.apply(func)
```

```python
#Sezonlara göre deplasman takımının ortalama perfect, good ve average değerlerinin gösterilmesi

pd.pivot_table(df2,'pts_away',index=["away"], columns=["season"], aggfunc=np.mean)
```
